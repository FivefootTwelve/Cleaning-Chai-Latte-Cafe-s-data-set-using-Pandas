```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/dirty_cafe_sales.csv')
```

```python
# --------
#CONVERSION OF QUANTITY COLUMN FROM OBJECT TO NUMERIC
# --------

df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
df['price_per_unit'] = pd.to_numeric(df['price_per_unit'], errors='coerce')
df['total_spent'] = pd.to_numeric(df['total_spent'], errors='coerce')

```

```python
# --------
# CONVERSION OF TRANSACTION DATE COLUMN
# --------

df['Transaction Date'] = pd.to_datetime(df['Transaction Date'], errors='coerce',format = 'mixed')
```

```python
# -------
# CHECKING FOR DUPLICATE ROWS
# -------

df.duplicated().value_counts()
```

```python
# ------
# CLEANING COLUMN NAMES
# ------

df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
# ------
# STANDARDIZE COLUMN TEXT VALUES i.e trailing spaces and conversion to lower case
# -------

df['item'] = df['item'].str.strip().str.lower()
df['location'] = df['location'].str.strip().str.lower()
df['payment_method'] = df['payment_method'].str.strip().str.lower()
```

```python
# ------
# HANDLING MISSING VALUES IN THE PRICE_PER_UNIT COLUMN
# ------

# Assign 1.0 to 'price_per_unit' for all 'cookie' items
df.loc[df['item'] == 'cookie', 'price_per_unit'] = 1.0

# Assign 1.5 to 'price_per_unit' for all 'tea' items
df.loc[df['item'] == 'tea', 'price_per_unit'] = 1.5

# Assign 2.0 to 'price_per_unit' for all 'coffee' items
df.loc[df['item'] == 'coffee', 'price_per_unit'] = 2.0

# Assign 3.0 to 'price_per_unit' for all 'cake' and 'juice' items
df.loc[df['item'].isin(['cake', 'juice']), 'price_per_unit'] = 3.0

# Assign 4.0 to 'price_per_unit' for all 'sandwich' and 'smoothie' items
df.loc[df['item'].isin(['sandwich','smoothie']), 'price_per_unit'] = 4.0

# Assign 5.0 to 'price_per_unit' for all 'salad' items
df.loc[df['item'] == 'salad', 'price_per_unit'] = 5.0

# Display a sample of the updated DataFrame to verify
display(df[df['item'] == 'sandwich'].head())
```

```python
# ------
# HANDLING MISSING VALUES IN THE QUANTITY COLUMN. THIS WILL BE DONE BY DIVIDING THE TOTAL SPENT BY THE PRICE PER UNIT
# ------

df.loc[
    (df['quantity'].isna()) &
    (df['price_per_unit'].notna())&
    (df['total_spent'].notna()), 'quantity'
] = df['total_spent']/df['price_per_unit']
```

```python
# ------
# HANDLING MISSING VALUES IN THE TOTAL SPENT COLUMN. THIS WILL BE DONE BY MULTIPLYING THE QUANTITY BY THE PRICE PER UNIT
# ------

df.loc[
    (df['total_spent'].isna())&
    (df['price_per_unit'].notna())&
    (df['quantity'].notna()), 'total_spent'
] = df['quantity']*df['price_per_unit']
```

```python
# ------
# HANDLING MISSING VALUES IN THE PRICE_PER_UNIT COLUMN. THIS WILL BE DONE BY DIVIDING THE TOTAL SPENT BY THE QUANTITY
# ------

df.loc[
    (df['price_per_unit']).isna()&
    (df['total_spent'].notna())&
    (df['quantity'].notna()), 'price_per_unit'
] = df['total_spent']/df['quantity']
```

```python
# ------
# HANDLING MISSING VALUES IN THE PRICE_PER_UNIT COLUMN
# ------

# Assign 1.0 to 'price_per_unit' for all 'cookie' items
df.loc[df['item'] == 'cookie', 'price_per_unit'] = 1.0

# Assign 1.5 to 'price_per_unit' for all 'tea' items
df.loc[df['item'] == 'tea', 'price_per_unit'] = 1.5

# Assign 2.0 to 'price_per_unit' for all 'coffee' items
df.loc[df['item'] == 'coffee', 'price_per_unit'] = 2.0

# Assign 3.0 to 'price_per_unit' for all 'cake' and 'juice' items
df.loc[df['item'].isin(['cake', 'juice']), 'price_per_unit'] = 3.0

# Assign 4.0 to 'price_per_unit' for all 'sandwich' and 'smoothie' items
df.loc[df['item'].isin(['sandwich','smoothie']), 'price_per_unit'] = 4.0

# Assign 5.0 to 'price_per_unit' for all 'salad' items
df.loc[df['item'] == 'salad', 'price_per_unit'] = 5.0
```

```python
# ------
# HANDLING MISSING VALUES IN THE ITEM COLUMN
# ------

df.loc[df['price_per_unit'] == 1.0, 'item'] = 'cookie'
df.loc[df['price_per_unit'] == 1.5, 'item'] = 'tea'
df.loc[df['price_per_unit'] == 2.0, 'item'] = 'coffee'
df.loc[df['price_per_unit'] == 5.0, 'item'] = 'salad'
```

```python
missing_items_df = df[df['item'].isnull()]
display(missing_items_df)
```

```python
# -------
# HANDLING NAN AND ERROR VALUES
# -------


df = df.replace(np.nan, 'unknown')
df = df.replace('error', 'unknown')
```
